```python
mov_30 = dict(np.load("/content/drive/My Drive/dynamic_pricing/results/moving_avg_30.npy", allow_pickle='TRUE').item())
mov_50 = dict(np.load("/content/drive/My Drive/dynamic_pricing/results/moving_avg_50.npy", allow_pickle='TRUE').item())
mov_100 = dict(np.load("/content/drive/My Drive/dynamic_pricing/results/moving_avg_100.npy", allow_pickle='TRUE').item())
```

```python

profit30 = np.array(list(mov_30['profit30']))[0]
proposed30 = np.array(list(mov_30['proposed30']))[0]
timestamp30 = np.array(list(mov_30['timestamp30']))[0]

profit50 = np.array(list(mov_50['profit50']))[0]
proposed50 = np.array(list(mov_50['proposed50']))[0]
timestamp50 = np.array(list(mov_50['timestamp50']))[0]

profit100 = np.array(list(mov_100['profit100']))[0]
proposed100 = np.array(list(mov_100['proposed100']))[0]
timestamp100 = np.array(list(mov_100['timestamp100']))[0]
```

```python
print(profit100.shape)
print(proposed100.shape)
print(timestamp100.shape)
```

```python
proposed_all = np.stack([proposed30,
  proposed50,
  proposed100], axis=0)

def smoothing_filter(array, kernel_length=100):
  return np.convolve(array, np.ones(kernel_length)/kernel_length, "same")

proposed_all.shape
```

```python


#plt.figure(dpi=300)
#smoothing filter
plt.rcParams['figure.figsize'] = [6, 3]
plt.rcParams["font.weight"] = "bold"

plt.plot([i for i in range(len(proposed30))], smoothing_filter(proposed30), label="mean")

plt.ylabel(f"Reward", fontweight='bold')
plt.xlabel("Time Step", fontweight='bold')
plt.show()
```

```python

#plt.figure(dpi=300)
#smoothing filter
plt.rcParams['figure.figsize'] = [6, 3]
plt.rcParams["font.weight"] = "bold"

plt.plot([i for i in range(len(proposed50))], smoothing_filter(proposed50), label="mean")

plt.ylabel(f"Reward", fontweight='bold')
plt.xlabel("Time Step", fontweight='bold')
plt.show()
```

```python

#plt.figure(dpi=300)
#smoothing filter
plt.rcParams['figure.figsize'] = [6, 3]
plt.rcParams["font.weight"] = "bold"

plt.plot([i for i in range(len(proposed100))], smoothing_filter(proposed100), label="mean")

plt.ylabel(f"Reward", fontweight='bold')
plt.xlabel("Time Step", fontweight='bold')
plt.show()
```
